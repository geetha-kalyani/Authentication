```python
import pandas as pd
import numpy as np
import math
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import seaborn
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
def concatDf(df,statsDf):
    for col in df.columns:
        statsDf[col]=df[col]
    return statsDf
```

```python
def calMagnitude(x,y,z):
    return math.sqrt(x*x+y*y+z*z)
```

```python
def create_df():
    return_df = pd.DataFrame()
    return return_df
```

```python
def read_file(filename):
    reading=create_df()
    reading=pd.read_csv(filename)
    return reading
```

```python
def calculating_square(data_frame):
#     working_df=create_df()
    data_frame['sqr_x'] = data_frame['Xvalue']**2
    data_frame['sqr_y'] = data_frame['Yvalue']**2
    data_frame['sqr_z'] = data_frame['Zvalue']**2
    return data_frame
```

```python
def calculating_mag(data_frame):
#     working_df=create_df()
    for i, data in data_frame.iterrows():
        data_frame.at[i,'magnitude'] = calMagnitude(data.Xvalue,data.Yvalue,data.Zvalue)
    return data_frame
```

```python
def calculating_power(data_frame):
    working_df=calculating_mean(calculating_square(data_frame))
#     print(working_df)
    working_df = working_df.rename(columns={"X_mean": "X_power", "Y_mean": "Y_power","Z_mean":"Z_power","MAG_mean":"MAG_power"})
#     print(working_df)
    return working_df
```

```python
 def calculating_mean(data_frame):     
    working_df =pd.DataFrame()
    i=0
    while i <len(data_frame):
        working_df=working_df.append(data_frame.iloc[i:i+100].mean(),ignore_index = True)
        i=i+50
    working_df = working_df.rename(columns={"Xvalue": "X_mean", "Yvalue": "Y_mean","Zvalue":"Z_mean","magnitude":"MAG_mean"})
#     print(data_frame)
    return working_df
```

```python
def calculating_var(data_frame):
    working_df=create_df()
    i=0
    while i <len(data_frame):
        working_df=working_df.append(data_frame.iloc[i:i+100].var(),ignore_index = True)
        i=i+50
    working_df = working_df.rename(columns={"Xvalue": "X_var", "Yvalue": "Y_var","Zvalue":"Z_var","magnitude":"MAG_var"})
    return working_df
```

```python
def calculating_std(data_frame):
    working_df=create_df()
    i=0
    while i <len(data_frame):
        working_df=working_df.append(data_frame.iloc[i:i+100].std(),ignore_index = True)
        i=i+50
    working_df = working_df.rename(columns={"Xvalue": "X_std", "Yvalue": "Y_std","Zvalue":"Z_std","magnitude":"MAG_std"})
    return working_df
```

```python
def calculating_min(data_frame):
    working_df=create_df()
    i=0
    while i <len(data_frame):
        working_df=working_df.append(data_frame.iloc[i:i+100].min(),ignore_index = True)
        i=i+50
    working_df = working_df.rename(columns={"Xvalue": "X_min", "Yvalue": "Y_min","Zvalue":"Z_min","magnitude":"MAG_min"})
    return working_df
```

```python
def calculating_max(data_frame):
    working_df=create_df()
    i=0
    while i <len(data_frame):
        working_df=working_df.append(data_frame.iloc[i:i+100].max(),ignore_index = True)
        i=i+50
    working_df = working_df.rename(columns={"Xvalue": "X_max", "Yvalue": "Y_max","Zvalue":"Z_max","magnitude":"MAG_max"})
    return working_df
```

```python
def calculating_RMS(data_frame):
    working_df=np.sqrt(calculating_power(data_frame))
    working_df.drop(['X_power','Y_power','Z_power','MAG_power'], axis=1, inplace=True)
    working_df = working_df.rename(columns={"sqr_x": "RMS_x", "sqr_y": "RMS_y","sqr_z":"RMS_z"})
    return working_df
```

```python
def feature_extraction(data_frame):
    data_frame1 = create_df()
    data_frame1 = calculating_mag(data_frame)
    compressed_data_frame = calculating_power(data_frame1)
#     compressed_data_frame = pd.merge(compressed_data_frame,calculating_mean(data_frame1))
#     compressed_data_frame = pd.concat([compressed_data_frame,calculating_var(data_frame1)],axis=1)
#     print(data_frame1)
#     data_frame1 = calculating_square(data_frame1)
#     compressed_data_frame = calculating_power(data_frame1)
    compressed_data_frame = pd.concat([compressed_data_frame,calculating_mean(data_frame1)],ignore_index = False,axis=1)
    compressed_data_frame = pd.concat([compressed_data_frame,calculating_var(data_frame1)],ignore_index = False, axis=1)
    compressed_data_frame = pd.concat([compressed_data_frame,calculating_std(data_frame1)],ignore_index = False, axis=1)
    compressed_data_frame = pd.concat([compressed_data_frame,calculating_min(data_frame1)],ignore_index = False, axis=1)
    compressed_data_frame = pd.concat([compressed_data_frame,calculating_max(data_frame1)],ignore_index = False, axis=1)
    compressed_data_frame = pd.concat([compressed_data_frame,calculating_RMS(data_frame1)],ignore_index = False, axis=1)
    compressed_data_frame.drop(['sqr_x','sqr_y','sqr_z'], axis=1, inplace=True)
#     compressed_data_frame = pd.concat([compressed_data_frame,calculating_square(data_frame1)],ignore_index = False, axis=1)
    return compressed_data_frame
```

```python
def final_func(i):
    stair = pd.read_csv(r'final\\'+str(i)+'_stair_Data.csv')
    walk = pd.read_csv(r'final\\'+str(i)+'_walk_Data.csv')
    stair_data=stair[["Xvalue","Yvalue","Zvalue"]]
    walk_data=walk[["Xvalue","Yvalue","Zvalue"]]
    final=feature_extraction(walk_data)
    return final
```
