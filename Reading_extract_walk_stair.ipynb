```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
import seaborn
```

```python
def reading_file(temp):
    path1='BB-MAS_Dataset'+'\\'+str(temp)+'\\'+str(temp)+'_HandPhone_Accelerometer_(Samsung_S6).csv'
    path2='BB-MAS_Dataset'+'\\'+str(temp)+'\\'+str(temp)+'_HandPhone_Checkpoints_(Samsung_S6).csv'
    try:
        handPhoneData = pd.read_csv(path1)
        handPhoneCheckPointData = pd.read_csv(path2)
    except:
        path1='BB-MAS_Dataset'+'\\'+str(temp)+'\\'+str(temp)+'_HandPhone_Accelerometer_(HTC_One).csv'
        path2='BB-MAS_Dataset'+'\\'+str(temp)+'\\'+str(temp)+'_HandPhone_Checkpoints_(HTC_One).csv'
        handPhoneData = pd.read_csv(path1)
        handPhoneCheckPointData = pd.read_csv(path2)
    return handPhoneData,handPhoneCheckPointData
```

```python
def print_df(data_frame):
    print(data_frame)
    
```

```python
def storing_walk(data_frame,num):
    path='final'+'\\'+str(num)+'_walk_Data.csv'
    data_frame.to_csv(path)
```

```python
def storing_stair(data_frame,num):
    path='final'+'\\'+str(num)+'_stair_Data.csv'
    data_frame.to_csv(path)
```

```python
def walk_data(num):
    handPhoneData,handPhoneCheckPointData=reading_file(num)
    walkTimings =[]
    timeLength = len(handPhoneCheckPointData.iloc[0]['time'])
    for i,data in handPhoneCheckPointData.iterrows():
        oneWalk=[]
        if(handPhoneCheckPointData.iloc[i]['eventType']=='DoorExit' and (i+1<len(handPhoneCheckPointData) and handPhoneCheckPointData.iloc[i+1]['eventType']=='DoorEntry' ) ):
            oneWalk.append(handPhoneCheckPointData.iloc[i]['time'][:(timeLength-4)])        
            oneWalk.append(handPhoneCheckPointData.iloc[i+1]['time'][:(timeLength-4)])
            walkTimings.append(oneWalk)
    walkIds = []
    for walks in walkTimings:
        frameList=[]
        for i,data in handPhoneData.iterrows():
            time =data.time
            if((time[:timeLength-4]) in walks):
                frameList.append(i)
        walkIds.append([frameList[0],frameList[len(frameList)-1]])
        
    walkDataFrame=pd.DataFrame()
    for l in walkIds:
        walkDataFrame = pd.concat([walkDataFrame,handPhoneData.iloc[l[0]:l[1]]])
    storing_walk(walkDataFrame,num)
    
```

```python
def stair_data(num):
    stairTimings =[]
    handPhoneData,handPhoneCheckPointData=reading_file(num);
    timeLength = len(handPhoneCheckPointData.iloc[0]['time'])
    for i,data in handPhoneCheckPointData.iterrows():
        onstair=[]
        if(handPhoneCheckPointData.iloc[i]['eventType']=='StairEntry' and (i+1<len(handPhoneCheckPointData) and handPhoneCheckPointData.iloc[i+1]['eventType']=='StairExit' ) ):
            onstair.append(handPhoneCheckPointData.iloc[i]['time'][:(timeLength-4)])        
            onstair.append(handPhoneCheckPointData.iloc[i+1]['time'][:(timeLength-4)])
            stairTimings.append(onstair)
    stairIds = []
    for stair in stairTimings:
        frameList=[]
        for i,data in handPhoneData.iterrows():
            time =data.time
            if((time[:timeLength-4]) in stair):
                frameList.append(i)
        stairIds.append([frameList[0],frameList[len(frameList)-1]])
    stairDataFrame = pd.DataFrame()
    for l in stairIds:
        stairDataFrame = pd.concat([stairDataFrame,handPhoneData.iloc[l[0]:l[1]]])
    storing_stair(stairDataFrame,num)
```

```python
for i in range(78,118):
    walk_data(i)
    stair_data(i)
    print(i)
```
